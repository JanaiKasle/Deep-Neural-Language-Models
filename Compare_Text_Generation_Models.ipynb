```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import numpy as np

# Accuracy data
cnn_accuracy = [0.3427, 0.4353, 0.5257, 0.5752, 0.5942]
rnn_accuracy = [0.5157, 0.5433, 0.5694, 0.5953, 0.6118]
lstm_accuracy = [0.5431, 0.5822, 0.6184, 0.6333, 0.6548]
gru_accuracy = [0.5799, 0.6113, 0.6368, 0.6580, 0.6881]

# Ground truth labels (0 or 1)
y_true = [0, 1, 1, 1, 1]

# Model predictions
y_pred_cnn = cnn_accuracy
y_pred_rnn = rnn_accuracy
y_pred_lstm = lstm_accuracy
y_pred_gru = gru_accuracy

# Calculate ROC curve
fpr_cnn, tpr_cnn, _ = roc_curve(y_true, y_pred_cnn)
fpr_rnn, tpr_rnn, _ = roc_curve(y_true, y_pred_rnn)
fpr_lstm, tpr_lstm, _ = roc_curve(y_true, y_pred_lstm)
fpr_gru, tpr_gru, _ = roc_curve(y_true, y_pred_gru)

# Calculate AUC (Area Under the Curve)
roc_auc_cnn = auc(fpr_cnn, tpr_cnn)
roc_auc_rnn = auc(fpr_rnn, tpr_rnn)
roc_auc_lstm = auc(fpr_lstm, tpr_lstm)
roc_auc_gru = auc(fpr_gru, tpr_gru)

# Plot ROC curve
plt.figure(figsize=(8, 4))
plt.plot(fpr_cnn, tpr_cnn, linestyle='-', marker='o', label=f'CNN (AUC = {roc_auc_cnn:.2f})')
plt.plot(fpr_rnn, tpr_rnn, linestyle='-', marker='s', label=f'RNN (AUC = {roc_auc_rnn:.2f})')
plt.plot(fpr_lstm, tpr_lstm, linestyle='-', marker='o', label=f'LSTM (AUC = {roc_auc_lstm:.2f})')
plt.plot(fpr_gru, tpr_gru, linestyle='-', marker='s', label=f'GRU (AUC = {roc_auc_gru:.2f})')

# Plotting the diagonal line representing random chance
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Chance')

# Customize the plot
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(True)
plt.show()
```
